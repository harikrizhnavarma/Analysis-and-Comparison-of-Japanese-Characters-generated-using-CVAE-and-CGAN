```python
file = open("CGAN_results.txt", 'r').read()
file = file.split("\n")
```

```python
gVals, dVals = [], []
for i, each in enumerate(file):
    dpattern = r'Loss D: ([0-9.]+)'
    gpattern = r'Loss G: ([0-9.]+)'
    dmatch = re.search(dpattern, each)
    gmatch = re.search(gpattern, each)
    dVals.append(float(dmatch.group(1)))
    gVals.append(float(gmatch.group(1)))
```

```python
plt.plot(gVals, label='Generator Loss')
plt.plot(dVals, label='Discriminator Loss')
plt.ylim(min(min(gVals), min(dVals)), max(max(gVals), max(dVals)))
plt.xlim(1, 500)
plt.legend()
plt.show()
```
